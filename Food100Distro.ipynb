```python
import glob
myPath = "/home/jupyter/currymuch/datasets/uecfood100-shorteats/images/"
counts = {}
for i in range (1,15):
    counter = len(glob.glob1(myPath+str(i),"*.jpg"))
    counts.append(counter)
    
print(counts)
```

```python
import os 

path = "/home/jupyter/currymuch/datasets/uecfood100-shorteats"
classfilename = 'food100.names'
classid2name = {}
classid = 0
file = os.path.join(path,classfilename)
if os.path.exists(file):
    with open(file) as cf:
        for line in cf.readlines():
            classname = line.strip('\n')
            classid2name[classname] = counts[classid]
            classid = classid + 1
            
print(classid2name)
```

```python
import matplotlib.pyplot as plt

def bar_chart(D):
    plt.bar(range(len(D)), list(D.values()), align='center')
    plt.xticks(range(len(D)), list(D.keys()))
    # # for python 2.x:
    # plt.bar(range(len(D)), D.values(), align='center')  # python 2.x
    # plt.xticks(range(len(D)), D.keys())  # in python 2.x

    plt.show()
```

```python
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}

matplotlib.rc('font', **font)
```

```python
plt.figure(figsize=(100,50)) 
plt.bar(*zip(*classid2name.items()))
plt.xticks(rotation=55, fontsize=40)
plt.yticks(fontsize=50)
plt.savefig('bar.png')
plt.show()
```
